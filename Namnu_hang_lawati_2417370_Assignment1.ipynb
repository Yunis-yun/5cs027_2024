# 3.2 Problem - 2 - Some Advance Data Exploration Task:

## Task - 1 - Setup Task - Preparing the South-Asia Dataset:

```python
#Task - 1:
import pandas as pd

south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India",
                        "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

# Load the DataFrame
df = pd.read_csv('/content/drive/MyDrive/dataset/WHR-2024-5CS037.csv') # Load the data into 'df'

# Filter the DataFrame
south_asia_df = df[df['Country name'].isin(south_asian_countries)]

# Save the filtered DataFrame to a CSV file
south_asia_df.to_csv('south_asia_happiness.csv', index=False)
print(south_asia_df)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Calculate Composite Score
south_asia_df['Composite Score'] = (0.40 * south_asia_df['Log GDP per capita'] +
                                    0.30 * south_asia_df['Social support'] +
                                    0.30 * south_asia_df['Healthy life expectancy'])

# Rank countries based on Composite Score
south_asia_df = south_asia_df.sort_values(by='Composite Score', ascending=False).reset_index(drop=True)

# Visualize top 5 countries
plt.figure(figsize=(8, 5))  # Slightly smaller figure size
sns.barplot(x='Composite Score', y='Country name', data=south_asia_df.head(5), palette="coolwarm")
plt.title('Top 5 South Asian Countries by Composite Score', fontsize=14)
plt.xlabel('Composite Score', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.tight_layout()
plt.show()

# Compare rankings with original Score
plt.figure(figsize=(10, 6))
plt.plot(south_asia_df['score'], label='Original Score', marker='o', linestyle='-', color='blue')
plt.plot(south_asia_df['Composite Score'], label='Composite Score', marker='s', linestyle='--', color='green')
plt.xticks(range(len(south_asia_df)), south_asia_df['Country name'], rotation=90)
plt.title('Comparison of Original and Composite Scores in South Asia', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
#Task - 3
# Calculate IQR for GDP per Capita and Score
Q1_gdp = south_asia_df['Log GDP per capita'].quantile(0.25)
Q3_gdp = south_asia_df['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

Q1_score = south_asia_df['score'].quantile(0.25)
Q3_score = south_asia_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

# Define outlier boundaries
lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

# Identify outliers
outliers = south_asia_df[
    (south_asia_df['Log GDP per capita'] < lower_bound_gdp) |
    (south_asia_df['Log GDP per capita'] > upper_bound_gdp) |
    (south_asia_df['score'] < lower_bound_score) |
    (south_asia_df['score'] > upper_bound_score)
]

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_df, hue=south_asia_df.index.isin(outliers.index), palette={True: 'red', False: 'blue'}) #Highlight outliers
plt.title('Outlier Detection in South Asian Countries')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()
#shows the outlier countries,log GDP per capita and score
print("Outlier Countries:")
print(outliers[['Country name', 'Log GDP per capita', 'score']])

# 4. Discuss the characteristics of these outliers and their potential impact on regional averages.

"""
Outliers are countries with either exceptionally high or low GDP per capita or happiness scores compared to others in the region.
These deviations can significantly affect the regional averages:

1. Countries with higher GDP per capita or scores may give a false impression that the region overall is wealthier or happier than it is.
2. Similarly, countries with extremely low values can make the region seem less prosperous or less happy than the actual average.

Outliers can also increase the perceived disparity among countries, highlighting significant economic or social inequalities within the region.
These extremes are often worth studying further as they may provide insights into unique circumstances influencing these metrics.
"""
```

## Task - 4 - Exploring Trends Across Metrics:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the South Asian countries data
south_asia_df = pd.read_csv('south_asia_happiness.csv')

# Calculate Pearson correlation for each metric and print it
metrics = ['Freedom to make life choices', 'Generosity']
for metric in metrics:
    correlation = south_asia_df['score'].corr(south_asia_df[metric])
    print(f"Pearson correlation between Score and {metric}: {correlation:.2f}")  # Rounded to 2 decimal places

# Create scatter plots with trendlines
for metric in metrics:
    sns.regplot(x=metric, y='score', data=south_asia_df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
    plt.title(f'Scatter Plot of Score vs. {metric}')
    plt.xlabel(metric)
    plt.ylabel('Score')
    plt.show()
```

```python
# Add GDP-Score Gap column
south_asia_df['GDP-Score Gap'] = south_asia_df['Log GDP per capita'] - south_asia_df['score']

# Rank by GDP-Score Gap
south_asia_df_ascending = south_asia_df.sort_values(by='GDP-Score Gap')
south_asia_df_descending = south_asia_df.sort_values(by='GDP-Score Gap', ascending=False)

# Highlight top 3 countries with largest positive and negative gaps
plt.figure(figsize=(12, 6))

# Top 3 positive gaps
plt.subplot(1, 2, 1)
sns.barplot(x='Country name', y='GDP-Score Gap', data=south_asia_df_descending.head(3), palette="viridis")
plt.xticks(rotation=45, ha='right')
plt.title('Top 3 Countries: Largest Positive GDP-Score Gap')

# Top 3 negative gaps
plt.subplot(1, 2, 2)
sns.barplot(x='Country name', y='GDP-Score Gap', data=south_asia_df_ascending.head(3), palette="viridis")
plt.xticks(rotation=45, ha='right')
plt.title('Top 3 Countries: Largest Negative GDP-Score Gap')

plt.tight_layout()
plt.show()

# Analyze reasons behind the gaps
print("\nAnalysis of GDP-Score Gaps:")
print("Countries with large positive gaps likely have high GDP but lower scores due to factors like social support, health, or freedom.")
print("Countries with large negative gaps might have lower GDP but higher social support, health, or freedom, leading to a higher happiness score.")
```

# 3.3 Problem - 3 - Comparative Analysis:

## Task - 1 - Setup Task - Preparing the Middle Eastern Dataset:

```python
# List of Middle Eastern countries
middle_east_countries = [
    "Bahrain", "Iran", "Iraq", "Israel", "Jordan", "Kuwait", "Lebanon", "Oman",
    "Palestine", "Qatar", "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"
]

# Filter the DataFrame for Middle Eastern countries
middle_east_df = df[df['Country name'].isin(middle_east_countries)]

# Calculate mean and standard deviation for Middle East scores
mean_score_middle_east = middle_east_df['score'].mean()
std_dev_score_middle_east = middle_east_df['score'].std()

# Calculate mean and standard deviation for South Asia (from previous task)
mean_score_south_asia = south_asia_df['score'].mean()
std_dev_score_south_asia = south_asia_df['score'].std()

# Print the results
print(f"Mean Happiness Score (Middle East): {mean_score_middle_east:.2f}")
print(f"Standard Deviation (Middle East): {std_dev_score_middle_east:.2f}")
print(f"Mean Happiness Score (South Asia): {mean_score_south_asia:.2f}")
print(f"Standard Deviation (South Asia): {std_dev_score_south_asia:.2f}")

# Compare the average happiness scores
if mean_score_middle_east > mean_score_south_asia:
    print("The Middle East has higher average happiness scores.")
elif mean_score_middle_east < mean_score_south_asia:
    print("South Asia has higher average happiness scores.")
else:
    print("Both regions have the same average happiness score.")
```

```python
def plot_top_bottom(df, region_name):
    """Plots the top and bottom 3 happiest countries in a given region."""

    # Sort the DataFrame by happiness score
    region_df_sorted = df.sort_values(by='score', ascending=False)

    # Get the top and bottom 3 countries
    top3 = region_df_sorted.head(3)
    bottom3 = region_df_sorted.tail(3)

    # Plotting the results
    plt.figure(figsize=(12, 6))

    #  3 happiest countries
    plt.subplot(1, 2, 1)
    sns.barplot(x='Country name', y='score', data=top3, palette="viridis")
    plt.xticks(rotation=45, ha='right')
    plt.title(f'Top 3 Happiest Countries in {region_name}')

    # Bottom 3 happiest countries
    plt.subplot(1, 2, 2)
    sns.barplot(x='Country name', y='score', data=bottom3, palette="viridis")
    plt.xticks(rotation=45, ha='right')
    plt.title(f'Bottom 3 Happiest Countries in {region_name}')

    plt.tight_layout()
    plt.show()

# Example usage (assuming 'middle_east_df' and 'south_asia_df' are already defined):
plot_top_bottom(middle_east_df, "Middle East")
plot_top_bottom(south_asia_df, "South Asia")
```

```python
# Task 3: Metric Comparisons

# Define the metrics for comparison
metrics = ['Log GDP per capita', 'Social support', 'Healthy life expectancy']

# Create grouped bar charts for each metric
for metric in metrics:
    plt.figure(figsize=(10, 6))

    # Create the grouped barplot
    sns.barplot(x='Country name', y=metric, data=middle_east_df, color='skyblue', label='Middle East')
    sns.barplot(x='Country name', y=metric, data=south_asia_df, color='orange', label='South Asia')

    plt.xticks(rotation=45, ha='right')
    plt.title(f'Comparison of {metric} between Middle East and South Asia')
    plt.xlabel('Country')
    plt.ylabel(metric)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Calculate and print the mean differences for each metric
for metric in metrics:
    diff_mean = abs(middle_east_df[metric].mean() - south_asia_df[metric].mean())
    print(f'Mean difference in {metric}: {diff_mean:.2f}')
```

```python
# Calculate the correlation of score with other metrics within each region
south_asia_corr = south_asia_df[['score', 'Freedom to make life choices', 'Generosity']].corr()  # Corrected variable name
middle_east_corr = middle_east_df[['score', 'Freedom to make life choices', 'Generosity']].corr()

# Plotting correlation heatmap for South Asia
plt.figure(figsize=(8, 6))
sns.heatmap(south_asia_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Analysis in South Asia')
plt.show()

# Plotting correlation heatmap for Middle East
plt.figure(figsize=(8, 6))
sns.heatmap(middle_east_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Analysis in Middle East')
plt.show()
```

```python
# Calculate IQR for South Asia and Middle East
south_asia_q1 = south_asia_df['score'].quantile(0.25)
south_asia_q3 = south_asia_df['score'].quantile(0.75)
south_asia_iqr = south_asia_q3 - south_asia_q1

middle_east_q1 = middle_east_df['score'].quantile(0.25)
middle_east_q3 = middle_east_df['score'].quantile(0.75)
middle_east_iqr = middle_east_q3 - middle_east_q1

# Detecting outliers using IQR (South Asia and Middle East)
south_asia_outliers = south_asia_df[(south_asia_df['score'] < (south_asia_q1 - 1.5 * south_asia_iqr)) |
                                    (south_asia_df['score'] > (south_asia_q3 + 1.5 * south_asia_iqr))]

middle_east_outliers = middle_east_df[(middle_east_df['score'] < (middle_east_q1 - 1.5 * middle_east_iqr)) |
                                      (middle_east_df['score'] > (middle_east_q3 + 1.5 * middle_east_iqr))]

# Plotting outliers for South Asia
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_outliers, color='red')
plt.title('South Asia: Outliers Based on Score and GDP per Capita')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()

# Plotting outliers for Middle East
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=middle_east_outliers, color='blue')
plt.title('Middle East: Outliers Based on Score and GDP per Capita')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()
```
